```python
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Load MNIST dataset
transform = transforms.Compose([transforms.ToTensor()])
mnist_data = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
mnist_loader = DataLoader(mnist_data, batch_size=1, shuffle=False)

# Find one image for each digit from 0 to 9
digit_samples = {}
for img, label in mnist_loader:
    label = label.item()
    if label not in digit_samples:
        digit_samples[label] = img
    if len(digit_samples) == 10: 
        break

plt.figure(figsize=(10, 2))
for i in range(10):
    img = digit_samples[i].squeeze()
    plt.subplot(1, 10, i + 1)
    plt.imshow(img, cmap='gray')
    plt.title(f"{i}")
    plt.axis('off')

plt.suptitle("Sample MNIST Images 0-9")
plt.show()
```

```python
import matplotlib.pyplot as plt
# LeNet Comparison accuracy:
# MNIST original, MNIST synthetic, MHIST original, MHIST synthetic
# 0.9927,0.8233,0.6725,0.6578
# Data
datasets = ["MNIST Original", "MNIST Synthetic", "MHIST Original", "MHIST Synthetic"]
accuracy = [0.9927, 0.8233, 0.6725, 0.6578]

# Colors: Blue for "Original", Orange for "Synthetic"
colors = ["#1f77b4", "#ff7f0e", "#1f77b4", "#ff7f0e"]

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(datasets, accuracy, color=colors)
plt.ylim(0, 1)
plt.ylabel("Accuracy")
# plt.title("LeNet Comparison Accuracy across Datasets")
plt.xticks(rotation=15)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Adding annotations for values and differences
for i, bar in enumerate(bars):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.02, f"{yval:.3f}", ha="center", va="bottom")

# Highlighting the difference in accuracy between Original and Synthetic within each dataset
mnist_diff = accuracy[0] - accuracy[1]
mhist_diff = accuracy[2] - accuracy[3]
plt.text(0.5, 0.5, f"MNIST Diff: {mnist_diff:.3f}", ha="center", color="blue", fontsize=12, weight="bold")
plt.text(2.5, 0.35, f"MHIST Diff: {mhist_diff:.3f}", ha="center", color="blue", fontsize=12, weight="bold")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
datasets = ["ConvNet", "LeNet", "PAD ConvNet", "DataDAM ConvNet", "DataDAM LeNet"]
accuracy = [0.9863, 0.9927, 0.9399, 0.8141, 0.9106]

# Define colors for original and synthetic groups
colors = ["orange", "lightgreen", "orange", "orange", "lightgreen"]

# Define x-axis positions with a gap
x_positions = [0, 1, 3, 4, 5]  # Creates a gap between Original and Synthetic bars

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(x_positions, accuracy, color=colors, width=0.8)
plt.xlabel("Original                                                                                      Synthetic            ")
plt.ylabel("Accuracy")
plt.ylim(0, 1)  # Set the y-axis limit to 0-1 for percentage clarity
plt.xticks(x_positions, datasets, rotation=0, fontsize = 8)

# Adding the accuracy values on top of each bar, formatted to 3 significant figures
for bar, value in zip(bars, accuracy):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 0.05, f"{value:.3f}",
             ha='center', va='bottom', color='black', fontsize=10)

# Adding a custom legend
custom_legend = [
    plt.Line2D([0], [0], color="orange", lw=4, label="Original ConvNet"),
    plt.Line2D([0], [0], color="lightgreen", lw=4, label="Original LeNet"),
    plt.Line2D([0], [0], color="orange", lw=4, label="Synthetic ConvNet", linestyle="", marker='//'),
    plt.Line2D([0], [0], color="lightgreen", lw=4, label="Synthetic LeNet", linestyle="", marker='//')
]
plt.legend(handles=custom_legend, loc="lower right")

plt.show()
```
